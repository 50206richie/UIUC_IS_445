```python
import pandas as pd
import altair as alt
import json
```

```python
license_data = 'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/licenses_fall2022.csv'
df = pd.read_csv(license_data)
df.head()
```

```python
license_counts = df.groupby('License Type').size().reset_index(name='Count')
# json_data1 = license_counts.to_json(orient='records')
# with open('is445hw8-1.json', 'w') as file:
#     file.write(json_data1)
license_counts.head()
```

```python
alt.Chart(license_counts).mark_bar().encode(
    x = 'License Type',
    y = alt.Y('Count', scale=alt.Scale(type='log'))
).properties(
    title='License Type Counts (Log Scale)'
)
# Get 'View Compiled Vega'
```
