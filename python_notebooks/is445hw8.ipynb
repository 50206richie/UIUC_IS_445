```python
import pandas as pd
import altair as alt
# import json
```

```python
license_data = 'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/licenses_fall2022.csv'
df = pd.read_csv(license_data)
df.head()
```

```python
license_counts = df.groupby('License Type').size().reset_index(name='Count')
# json_data1 = license_counts.to_json(orient='records')
# with open('is445hw8-1.json', 'w') as file:
#     file.write(json_data1)
license_counts.head()
```

```python
alt.Chart(license_counts).mark_bar().encode(
    x = 'License Type',
    y = alt.Y('Count', scale=alt.Scale(type='log'))
).properties(
    title='License Type Counts (Log Scale)'
)
# Get 'View Compiled Vega'
```

```python
df['Effective Date'] = pd.to_datetime(df['Effective Date'], format='%m/%d/%Y', errors='coerce')
df['Expiration Date'] = pd.to_datetime(df['Expiration Date'], format='%m/%d/%Y', errors='coerce')
df['Effective Duration (Days)'] = (df['Expiration Date'] - df['Effective Date']).dt.days

df_plot2 = df[~df['Effective Date'].isna() & ~df['Expiration Date'].isna()]
# license_duration[['Effective Duration']]

license_duration = (
                    df_plot2.groupby('License Type')['Effective Duration (Days)']
                    .mean().reset_index().sort_values('Effective Duration (Days)')
                   ).rename(columns={'Effective Duration (Days)': 'Average Effective Duration (Days)'})
license_duration['Average Effective Duration (Days)'] = license_duration['Average Effective Duration (Days)'].round(2)

license_duration
```

```python
alt.Chart(license_duration).mark_bar().encode(
    x=alt.X('License Type', sort='y'),
    y='Average Effective Duration (Days)',
    color=alt.Color(
        'Average Effective Duration (Days)', 
        scale=alt.Scale(range=['lightblue', 'darkblue'])),
    tooltip=['License Type', alt.Tooltip('Average Effective Duration (Days)', format='.2f')]
).properties(
    title='Average Duration of all License Types'
)
```
